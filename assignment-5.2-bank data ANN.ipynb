# splitting into x and y

```python

x = dataset.iloc[:,:4].values
y = dataset.iloc[:,4:5].values
```

# Handling categorical data

```python
dataset["State"].unique()
```

```python
# Arrangement: California, Florida, New York
x.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct=ColumnTransformer([("one", OneHotEncoder(), [3])], remainder="passthrough")
x=ct.fit_transform(x)
```

# splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python

print("x_train shape", "\t", x_train.shape)
print("x_test shape", "\t", x_test.shape)
print("y_train shape", "\t", y_train.shape)
print("y_test shape", "\t", y_test.shape)
print("x shape", "\t", x.shape)
print("y shape", "\t", y.shape)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

# Model Building

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
regression = Sequential()
```

```python
x.shape
```

#### Adding Input Layer

```python
regression.add(Dense( units=6 , kernel_initializer= "random_uniform", activation="relu"))
```

#### Adding Hidden Layer

```python
regression.add(Dense( units= 12, kernel_initializer= "random_uniform", activation="relu"))
```

#### Adding Output Layer

```python
regression.add(Dense( units= 1, kernel_initializer= "random_uniform", activation="relu"))
```

#### Compile data

```python
regression.compile(optimizer = "adam", loss = "mse", metrics=['mse'])
```

#### Fit Model

```python
regression.fit(x_train,y_train, epochs=10000)
```

```python
ypred = regression.predict(x_test)
```
